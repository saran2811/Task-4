###### merge those two csv files (after getting as dataframes, get them as a single dataframe)

```python
df=pd.read_csv("/content/college_1.csv")
df.head()
```

```python
df1=pd.read_csv("/content/college_2.csv")
df1.head()
```

```python
df_merge=pd.merge(df,df1,how='outer')
```

```python
df_merge
```

```python
df_merge.fillna(0,inplace=True)
```

```python
import random
for i in range(len(df_merge['CodeKata Score'])):
    if df_merge['CodeKata Score'][i] in range(10000,25000):
        if df_merge['python'][i]<80:
           t=random.uniform(65,95)
           df_merge['python'][i]=t
        if df_merge['mysql'][i]<80:
           t=random.uniform(65,95)
           df_merge['mysql'][i]=t
        if df_merge['python_en'][i]<80:
           t=random.uniform(65,95)
           df_merge['python_en'][i]=t
        if df_merge['computational_thinking'][i]<80:
           t=random.uniform(65,95)
           df_merge['computational_thinking'][i]=t
    if df_merge['CodeKata Score'][i] in range(5000,10000):
        if df_merge['python'][i]<65:
           t=random.uniform(40,65)
           df_merge['python'][i]=t
        if df_merge['mysql'][i]<80:
           t=random.uniform(35,65)
           df_merge['mysql'][i]=t
        if df_merge['python_en'][i]<80:
           t=random.uniform(35,65)
           df_merge['python_en'][i]=t
        if df_merge['computational_thinking'][i]<80:
           t=random.uniform(35,65)
           df_merge['computational_thinking'][i]=t
    if df_merge['CodeKata Score'][i] in range(0,5000):
        if df_merge['python'][i]<65:
           t=random.uniform(10,35)
           df_merge['python'][i]=t
        if df_merge['mysql'][i]<80:
           t=random.uniform(10,35)
           df_merge['mysql'][i]=t
        if df_merge['python_en'][i]<80:
           t=random.uniform(10,35)
           df_merge['python_en'][i]=t
        if df_merge['computational_thinking'][i]<80:
           t=random.uniform(10,35)
           df_merge['computational_thinking'][i]=t
df_merge
```
